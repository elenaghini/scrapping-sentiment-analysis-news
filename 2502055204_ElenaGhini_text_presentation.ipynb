```python
import numpy as np

def document_vector(word_vectors, doc):
    # Get word vectors for words in the document
    doc_vectors = [word_vectors[word] for word in doc if word in word_vectors]

    # Return the average vector for the document
    return np.mean(doc_vectors, axis=0)

# Example: Convert a list of documents to vectors
document_vectors = [document_vector(model_cbow.wv, doc) for doc in word_token]
```

```python
#spliting data
from sklearn.model_selection import train_test_split

#X_train, X_test, y_train, y_test = train_test_split(document_vectors, Label, test_size=0.2, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(document_vectors, Label, test_size=0.3, random_state=50)


#spliting data
#from sklearn.model_selection import train_test_split
#X_train, X_test, y_train, y_test= train_test_split(document_vectors['Label', test_size=0.2, random_state=42])
```

```python
from sklearn.ensemble import RandomForestClassifier

# Initialize the model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the model
rf_model.fit(X_train, y_train)

# Predict on the test set
rf_predictions = rf_model.predict(X_test)
```

**Pemodelan RandomForest vs SVM**

```python
from sklearn.metrics import accuracy_score, classification_report

# Evaluate accuracy
accuracy = accuracy_score(y_test, rf_predictions)
print(f"Random Forest Accuracy: {accuracy:.2f}")

# Display classification report
print("Classification Report:")
print(classification_report(y_test, rf_predictions))
```

```python
from sklearn.svm import SVC

# Initialize the model
svm_model = SVC(kernel='linear', C=1.0, random_state=42)

# Train the model
svm_model.fit(X_train, y_train)

# Predict on the test set
svm_predictions = svm_model.predict(X_test)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report

# Initialize the SVM model with a polynomial kernel
svm_model_poly = SVC(kernel='poly', degree=3, C=1.0, random_state=42)

# Train the model
svm_model_poly.fit(X_train, y_train)

# Predict on the test set
svm_predictions_poly = svm_model_poly.predict(X_test)

# Evaluate accuracy
accuracy_poly = accuracy_score(y_test, svm_predictions_poly)
print(f"SVM with Polynomial Kernel (Degree 3) Accuracy: {accuracy_poly:.2f}")

# Display classification report
print("Classification Report:")
print(classification_report(y_test, svm_predictions_poly))

```
